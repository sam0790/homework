```python
def reverse(n = 11111):
    n = str(n)
    lis = []
    n1 = ''
    for i in n:
        lis.append(i)
        lis.sort(reverse=True) 
        
    for i in lis:
        n1 = n1 + i
        
    return n1   
print(reverse(43905)) 
print(reverse(584366))
```
