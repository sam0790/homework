```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)

print(purchase_data)
```

## Player Count

* Display the total number of players


```python
unique_player = purchase_data["SN"].unique()
Players = len(unique_player)


total_player = pd.Series(data=Players , name="Total Players")
total_player
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
uni_item = purchase_data["Item Name"].unique()
unique_item = len(uni_item)
#print(unique_item)

calc_price = purchase_data["Price"]

average_price = (calc_price.sum() / len(calc_price))
ave_price = round(average_price,2)
#print(average_price)


number_purchase = len(purchase_data["Purchase ID"])
#print(number_purchase)


total_revenue = purchase_data["Price"].sum()
#print(total_revenue)

dict = {"Number of Unique Items" : unique_item ,"Average Price": ave_price , "Number of Purchases" : number_purchase , "Total Revenue" : total_revenue }
#print(dict)

summary = pd.DataFrame(data=dict, index=[0])
summary["Average Price"] = summary["Average Price"].apply(str)
summary["Average Price"] = "$" + summary["Average Price"]
summary["Total Revenue"] = summary["Total Revenue"].apply(str)
summary["Total Revenue"] = "$" + summary["Total Revenue"]

summary
```

## Gender Demographics


* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
df_male = (purchase_data["Gender"] == "Male")
total_male = len(purchase_data[df_male])
#print(total_male)

df_female = (purchase_data["Gender"] == "Female")
total_female = len(purchase_data[df_female])
#print(total_female)

df_other = (purchase_data["Gender"] == "Other / Non-Disclosed")
total_other = len(purchase_data[df_other])
#print(total_other)

male_perc = (total_male / 780)*100
male_per = round(male_perc,2)
#print(male_per)

female_perc = (total_female / 780)*100
female_per = round(female_perc,2)
#print(female_perc)

other_perc = (total_other / 780)*100
other_per = round(other_perc,2)
#print(other_perc)

list1 = [total_male,total_female,total_other]
#print(list1)

list2 = [male_per,female_per,other_per]
#print(list2)

dict = {"Total Count" : list1 , "Percentage of Players" : list2}

count_df = pd.DataFrame(data=dict, index=["Male","Female","Other / Non-Disclosed"])
count_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
group_by = purchase_data.groupby("Gender")
group_by.count()

pur_id = group_by["Purchase ID"].count()
#print(pur_id)

avg_pur_price = group_by["Price"].mean()
avg_price = round(avg_pur_price,2)
#print(avg_pur_price)

total_val = group_by["Price"].sum()
tot_val = round(total_val,2)
#print(total_val)

#sum_total = (group_by["Price"].sum())
#print(sum_total)

list = {"Purchase Count":pur_id , "Average Purchase Price" : avg_price, "Total Purchase Value": tot_val}

summary = pd.DataFrame(data=list)
summary["Average Purchase Price"] = summary["Average Purchase Price"].apply(str)
summary["Average Purchase Price"] = "$" + summary["Average Purchase Price"]
summary["Total Purchase Value"] = summary["Total Purchase Value"].apply(str)
summary["Total Purchase Value"] = "$" + summary["Total Purchase Value"]

summary
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
age_bins = [0,9,14,19,24,29,34,39,100]
bins_lab = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]
pd.cut(purchase_data["Age"], age_bins, labels=bins_lab)

purchase_data["age_data"] = pd.cut(purchase_data["Age"], age_bins, labels=bins_lab)
purchase_data

group_age = purchase_data.groupby("age_data")

total_count = group_age["Purchase ID"].count()
#print(total_count)

perc_count = (total_count/780)*100
perc_countr = round(perc_count,2)
#print(perc_countr)

dict = {"Total Count": total_count,"Percentage of Players":perc_countr}

summary = pd.DataFrame(data=dict)
summary
```
