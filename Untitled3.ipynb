```python
import pandas as pd

belize = pd.read_csv("belize_airbnb_listings.csv")

bel_df = pd.DataFrame(belize)
bel_df
```

```python
new_bel = bel_df.groupby("neighbourhood")

new_bel.head()
```

```python
meandf = new_bel['price'].mean()
total = new_bel['number_of_reviews'].sum()
ave_avail = new_bel["availability_365"].mean()

meandf.head()
```

```python
total.head()
```

```python
ave_avail.head()
```

```python
new_dict = {"Price": meandf, "Number of reviews": total, "Availability": ave_avail }

new_df = pd.DataFrame(new_dict)
new_df.index_reset()
new_df
```
