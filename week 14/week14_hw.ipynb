### 1.	Perform pre-processing on the full dataset here: https://www.kaggle.com/sohier/calcofi

### 2.	Perform feature selection (decide what columns should be included in your analysis). How did you figure out what features were important?


Based on RFE feature selection and selecting number of features 5 because we will be using 5 features to predict temperature of water, we can see from the ranking list below, columns Depthm,Salnty,R_Depth,R_SVA and R_PRES are the features we need to predict target variable T_degC. 

As you can see you in the ranking list below you get 1 as your first choices for the features for target variable temperature. And based on this result we have selected these features.


Selected Features (columns) : Depthm , Salnty , R_Depth , R_SVA , R_PRES

```python
from sklearn.feature_selection import RFE

X = df10.drop('T_degC',axis=1)
y = df10['T_degC']

reg = LinearRegression()
rfe = RFE(reg,n_features_to_select=5)
rfe_ = rfe.fit(X, y)

print(X.columns)
print(rfe_.support_)
print(rfe.ranking_)
```

### 3.	Perform multiple linear regression on the dataset. Be sure to display the final resulting equation with the coefficients and intercept in markdown. 

```python
reg = LinearRegression()

X = df10.drop(['STheta','T_degC','O2Sat','RecInd','R_DYNHT'],axis=1)
y = df10['T_degC']

#print(X.columns)
#print(y)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,  random_state=45)

reg.fit(X_train,y_train)

accuracy_score = reg.score(X_test, y_test)

print(accuracy_score)
```

```python
print(reg.intercept_)
print(reg.coef_)
```

```python
dia_df = pd.read_csv('../diabetes.csv')

dia_df
```

```python

X = dia_df.drop('Outcome',axis=1)
y = dia_df['Outcome']

reg = LinearRegression()
rfe = RFE(reg,n_features_to_select=3)
rfe_ = rfe.fit(X, y)

print(X.columns)
print(rfe_.support_)
print(rfe.ranking_)

#Based on the ranking list from below we will be using BMI and DiabetesPedigreeFunction features as our explanatory
#variables.
```

```python
X = dia_df.drop('Outcome', axis=1)
y = dia_df['Outcome']

# Split into training and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state=42, stratify=y)

#Standardize
sc= StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)
```

```python
log_reg = LogisticRegression()

# fit the model with data
log_reg.fit(X_train,y_train)

#
y_pred = log_reg.predict(X_test)

score = log_reg.score(X,y)
print(score)
```

```python
conf_matrix = confusion_matrix(y_test, y_pred)
conf_matrix
```

```python
print(classification_report(y_test, y_pred))
```

```python
print("Accuracy :",metrics.accuracy_score(y_test, y_pred))
print("Precision :",metrics.precision_score(y_test, y_pred))
print("Recall :",metrics.recall_score(y_test, y_pred))
```

```python
from statsmodels.formula.api import logit

bmi_ped_int=logit("Outcome ~ BMI * DiabetesPedigreeFunction", data=dia_df).fit()
print(bmi_ped_int.params)
```
