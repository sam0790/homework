### KNN

K-Nearest Neighbors also known as KNN is theory which assumes that similar things exist in close proximity. In simple words, similar things are near to each other. KNN captures the idea of similarity of distance, proximity or closeness. KNN algorithm is as simple, easy to implement supervised machine learning algorithm that can be used to solve both classification and regression problems.


```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

diabetes_df = pd.read_csv("diabetes.csv")
diabetes_df
```

```python
X = diabetes_df.drop('Outcome', axis=1)
y = diabetes_df['Outcome']

#split data into training and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=45)

#Standardization
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)

knn = KNeighborsClassifier(n_neighbors=119)

knn.fit(X,y)
y_predicted = knn.predict(X)
print(y_predicted)
print(knn.score(X,y))
```

```python
def find_k(X,y,X_train,y_train,X_test):
    
    for i in range(120):
        score = 0
        k = i
        if i%2 == 0:
            continue
        if i%2 != 0:
            knn = KNeighborsClassifier(n_neighbors=i)
            knn.fit(X,y)
            y_predicted = knn.predict(X)
            score1 = knn.score(X,y)
            if score1 > score:
                    score = score1
                    
            elif score1 < score:
                break
       
        
    return (k, score1)

find_k(X,y,X_train,y_train,X_test)
    
```
